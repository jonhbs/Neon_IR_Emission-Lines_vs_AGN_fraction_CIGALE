```python
from astropy.io import ascii
from astropy.table import Table, Column
from astropy.coordinates import SkyCoord

import astropy.units as u
import numpy as np
```

```python
path_FINAL = '../Datos/Final/'
```

```python
Ne_Sy = Table.read(path_FINAL+'final_results.fits',format='fits')
c = SkyCoord(ra=Ne_Sy['RA']*u.degree, dec=Ne_Sy['DEC']*u.degree, frame='icrs')
Ne_Sy['Coordinates (ICRS)'] = c.to_string('hmsdms',precision=2)
Ne_Sy.rename_column('Main_id','Main Id')
Ne_Sy.rename_column('redshift','Redshift')
Ne_Sy.rename_column('otype','Object Type (SMB)')
Ne_Sy.rename_column('CountPoints','No. Phot Points')
```

```python
# Adding units
Ne_Sy['RA'] = Ne_Sy['RA']*u.deg
Ne_Sy['DEC'] = Ne_Sy['DEC']*u.deg
Ne_Sy['[NeV]*'] = Ne_Sy['[NeV]*']*u.Jy
Ne_Sy['[NeV]_err*'] = Ne_Sy['[NeV]_err*']*u.Jy
Ne_Sy['[NeII]*'] = Ne_Sy['[NeII]*']*u.Jy
Ne_Sy['[NeII]_err*'] = Ne_Sy['[NeII]_err*']*u.Jy
Ne_Sy[r'[NeV]$^+$'] = Ne_Sy[r'[NeV]$^+$']*u.Jy
Ne_Sy[r'[NeV]_err$^+$'] = Ne_Sy[r'[NeV]_err$^+$']*u.Jy
Ne_Sy[r'[NeIII]$^+$'] = Ne_Sy[r'[NeIII]$^+$']*u.Jy
Ne_Sy[r'[NeIII]_err$^+$'] = Ne_Sy[r'[NeIII]_err$^+$']*u.Jy
Ne_Sy[r'[NeIII]$^-$'] = Ne_Sy[r'[NeIII]$^-$']*u.Jy
Ne_Sy[r'[NeIII]_err$^-$'] = Ne_Sy[r'[NeIII]_err$^-$']*u.Jy
Ne_Sy[r'[NeII]$^-$'] = Ne_Sy[r'[NeII]$^-$']*u.Jy
Ne_Sy[r'[NeII]_err$^-$'] = Ne_Sy[r'[NeII]_err$^-$']*u.Jy
# Changing print format
Ne_Sy['Redshift'].info.format = '.4f'
#Ne_Sy['[NeV]*'].info.format = '.2e'
#Ne_Sy['[NeV]_err*'].info.format = '.2e'
#Ne_Sy['[NeII]*'].info.format = '.2e'
#Ne_Sy['[NeII]_err*'].info.format = '.2e'
#Ne_Sy[r'[NeV]$^+$'].info.format = '.2e'
#Ne_Sy[r'[NeV]_err$^+$'].info.format = '.2e'
#Ne_Sy[r'[NeIII]$^+$'].info.format = '.2e'
#Ne_Sy[r'[NeIII]_err$^+$'].info.format = '.2e'
#Ne_Sy[r'[NeIII]$^-$'].info.format = '.2e'
#Ne_Sy[r'[NeIII]_err$^-$'].info.format = '.2e'
#Ne_Sy[r'[NeII]$^-$'].info.format = '.2e'
#Ne_Sy[r'[NeII]_err$^-$'].info.format = '.2e'
#Ne_Sy['log_[NeV]/[NeII]'].info.format = '.3f'
#Ne_Sy['log_[NeV]/[NeII]_err'].info.format = '.3f'
#Ne_Sy['log_[NeV]/[NeIII]'].info.format = '.3f'
#Ne_Sy['log_[NeV]/[NeIII]_err'].info.format = '.3f'
#Ne_Sy['log_[NeIII]/[NeII]'].info.format = '.3f'
#Ne_Sy['log_[NeIII]/[NeII]_err'].info.format = '.3f'
print('Total number of galaxies with Ne ratios data: ',len(Ne_Sy))
print('Colmunas: ',Ne_Sy.columns)
```

```python
sample = Table(Ne_Sy)
sample = sample['Main Id','Redshift','Coordinates (ICRS)','Object Type (SMB)','No. Phot Points']
```

```python
r1_Sy = Table(Ne_Sy)
r1_Sy.remove_rows(np.where(np.isnan(r1_Sy[r'[NeV]/[NeII]*']))[0])
r1_Sy = r1_Sy['Main Id','Object Type (SMB)','[NeV]*',
              '[NeV]_err*','[NeII]*','[NeII]_err*',
              'log_[NeV]/[NeII]','log_[NeV]/[NeII]_err']
r1_Sy.rename_column('[NeV]*','[NeV]')
r1_Sy.rename_column('[NeV]_err*','[NeV] err')
r1_Sy.rename_column('[NeII]*','[NeII]')
r1_Sy.rename_column('[NeII]_err*','[NeII] err')

r1_Sy['[NeV]'] = Column(['{:.2e}'.format(x) for x in r1_Sy['[NeV]']], name='[NeV]')
r1_Sy['[NeV] err'] = Column(['{:.2e}'.format(x) for x in r1_Sy['[NeV] err']], name='[NeV] err')
r1_Sy['[NeII]'] = Column(['{:.2e}'.format(x) for x in r1_Sy['[NeII]']], name='[NeII]')
r1_Sy['[NeII] err'] = Column(['{:.2e}'.format(x) for x in r1_Sy['[NeII] err']], name='[NeII] err')
r1_Sy['log_[NeV]/[NeII]'] = Column(['{:.3f}'.format(x) for x in r1_Sy['log_[NeV]/[NeII]']], name='log_[NeV]/[NeII]')
r1_Sy['log_[NeV]/[NeII]_err'] = Column(['{:.3f}'.format(x) for x in r1_Sy['log_[NeV]/[NeII]_err']], name='log_[NeV]/[NeII]_err')

NeV=[]
NeII=[]
log_r1 = []
for i in range(len(r1_Sy)):
    NeV.append(r1_Sy['[NeV]'][i]+r' ($\pm$'+r1_Sy['[NeV] err'][i]+')')
    NeII.append(r1_Sy['[NeII]'][i]+r' ($\pm$'+r1_Sy['[NeII] err'][i]+')')
    log_r1.append(r1_Sy['log_[NeV]/[NeII]'][i]+r' ($\pm$'+r1_Sy['log_[NeV]/[NeII]_err'][i]+')')

r1_Sy['[Ne V]'] = NeV
r1_Sy['[Ne II]'] = NeII
r1_Sy[r'$\log(\frac{[Ne V]}{[Ne II]})$'] = log_r1
```
